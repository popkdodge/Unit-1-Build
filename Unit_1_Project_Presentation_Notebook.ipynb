The data is source from [lazyfa.com](https://www.lazyfa.com/) , [Nasdaq](https://www.nasdaq.com/), and [US Department of Tranportation](https://www.transportation.gov/).

The data from lazfa.com is stock fundamental data such as a company's balance sheet, income statements, cash flow statements, and even in term of stock matrics. This website offer the option for its user to download a stock fundamental in CSV format. Its user may choose from dowloading its data in Quarteries, Annualies, or TTM (Trailing Twelth Month) TTM take the Quarties and combine data for the last 3 trailing Quarter to create an Annualies like statements every quarter.

The data from Nasdaq is historical stock prices. The website offer anywhere from the last month to the last 5 years in dalies formats.

The data from DOT is just the percent of marketshare Airlines companies have.

## Cleaning the data
---

#### Switching the columns and setting the index.

````
LUV_Q = LUV_Q.set_index('Reporting Period')
LUV_Q = LUV_Q.T
LUV_Q
````

#### Creating a Ticker Columns with approprite Ticker Symbol and moving the column to the front.

````
LUV_Q['Ticker'] = 'LUV'
Ticker = LUV_Q['Ticker']
LUV_Q.drop(labels=['Ticker'], axis=1,inplace = True)
LUV_Q.insert(0, 'Ticker', Ticker)
LUV_Q
````

#### Reseting the Index , Dropping NaNs Columns and renaming some columns.

````
LUV_Q = LUV_Q.reset_index()
LUV_Q = LUV_Q.drop(['Average Equity','Return on Sales','Asset Turnover','Average Assets','Invested Capital Average',
                     'Return on Invested Capital','Return on Average Assets', 'Return on Average Equity'],axis=1)
LUV_Q.columns =['Time', 'Ticker', 'Book Value per Share', 'Current Ratio',
       'Debt to Equity Ratio', 'Dividend Yield', 'EBITDA', 'EBITDA [USD]',
       'EBITDA Margin', 'Earnings before Tax', 'Enterprise Value',
       'Enterprise Value over EBIT', 'Enterprise Value over EBITDA',
       'Free Cash Flow', 'Free Cash Flow per Share',
       'Foreign Currency to USD Exchange Rate', 'Gross Margin',
       'Invested Capital', 'Market Capitalization', 'Profit Margin',
       'Price to Earnings (Damodaran Method)', 'Price to Earnings Ratio',
       'Price to Sales Ratio', 'Price to Sales (Damodaran Method)',
       'Price to Book Value', 'Sales per Share', 'Payout Ratio',
       'Tangible Asset Value', 'Tangible Asset Book Value per Share',
       'Working Capital']
LUV_Q
````

````
DAL_Q  = pd.read_csv('https://raw.githubusercontent.com/popkdodge/Unit-1-Build/master/Data_set_csv/DAL_Quarterlies.csv')
DAL_Q = DAL_Q.set_index('Reporting Period')
DAL_Q = DAL_Q.T
DAL_Q['Ticker'] = 'DAL_Q'
Ticker = DAL_Q['Ticker']
DAL_Q.drop(labels=['Ticker'], axis=1,inplace = True)
DAL_Q.insert(0, 'Ticker', Ticker)
DAL_Q = DAL_Q.reset_index()
DAL_Q = DAL_Q.drop(['Average Equity','Return on Sales','Asset Turnover','Average Assets','Invested Capital Average',
                     'Return on Invested Capital','Return on Average Assets', 'Return on Average Equity'],axis=1)
DAL_Q.columns =['Time', 'Ticker', 'Book Value per Share', 'Current Ratio',
       'Debt to Equity Ratio', 'Dividend Yield', 'EBITDA', 'EBITDA [USD]',
       'EBITDA Margin', 'Earnings before Tax', 'Enterprise Value',
       'Enterprise Value over EBIT', 'Enterprise Value over EBITDA',
       'Free Cash Flow', 'Free Cash Flow per Share',
       'Foreign Currency to USD Exchange Rate', 'Gross Margin',
       'Invested Capital', 'Market Capitalization', 'Profit Margin',
       'Price to Earnings (Damodaran Method)', 'Price to Earnings Ratio',
       'Price to Sales Ratio', 'Price to Sales (Damodaran Method)',
       'Price to Book Value', 'Sales per Share', 'Payout Ratio',
       'Tangible Asset Value', 'Tangible Asset Book Value per Share',
       'Working Capital']
DAL_Q
````

````
AAL_Quote = pd.read_csv('https://raw.githubusercontent.com/popkdodge/Unit-1-Build/master/Data_set_csv/AAL_Quote.csv')
AAL_Quote[' Open'] = AAL_Quote[' Open'].str.replace('$', '')
AAL_Quote[' High'] = AAL_Quote[' High'].str.replace('$', '')
AAL_Quote[' Low'] = AAL_Quote[' Low'].str.replace('$', '')
AAL_Quote[' Close/Last'] = AAL_Quote[' Close/Last'].str.replace('$', '')
AAL_Quote[' Open'] = AAL_Quote[' Open'].astype(float)
AAL_Quote[' High'] = AAL_Quote[' High'].astype(float)
AAL_Quote[' Low'] = AAL_Quote[' Low'].astype(float)
AAL_Quote[' Close/Last'] = AAL_Quote[' Close/Last'].astype(float)
AAL_Quote['Date'] = pd.to_datetime(AAL_Quote['Date'], infer_datetime_format=True)
AAL_Quote['Ticker'] = 'AAL'
Ticker = AAL_Quote['Ticker']
AAL_Quote.drop(labels=['Ticker'], axis=1,inplace = True)
AAL_Quote.insert(0, 'Ticker', Ticker)
AAL_Quote
````

##Visualization code.

````
import pandas as pd
import plotly.graph_objects as go
import matplotlib.pyplot as plt

MarketShare = {'Airlines': ['Delta', 'American', 'Southwest', 'United', 'Alaska', 'JetBlue', 'Spirit', 'Frontier', 'Skywest', 'Hawaiian', 'Other' ], 
               'Marketshare %': ['17.6','17.5','16.7','14.8','6.4','5.5','4.3','3.6','2.9','1.6','9.6']}
MarketShareP = pd.DataFrame(data=MarketShare)
labels = MarketShareP['Airlines']
values = MarketShareP['Marketshare %']
fig = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='label+percent'
                            )])
fig.update_layout(
title={
        'text': "Marketshare of US airlines companies",
        'y':0.9,
        'x':0.48,
        'xanchor': 'center',
        'yanchor': 'top',
        'font': dict(
        family="Courier New, monospace",
        size=23,
        color="#212121")},
annotations=[
             dict(
            x=.47,
            y=-0.21,
            showarrow=False,
            text="Source: transtats.bts.gov - United States Department of Transportation",
            xref="paper",
            yref="paper"
        )
]
        )

fig.show()
````
